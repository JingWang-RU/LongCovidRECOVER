```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

def build_fixed_length_matrix(
    df: pd.DataFrame,
    id_col: str = "PARTICIPANT_ID",
    filter_col: str = "CONCEPT_NAME",
    filter_value = topc[0],
    value_col: str = "value"  # the column you want to cluster on
):
# if True:
    sub = df.loc[df[filter_col] ==filter_value].copy()
    counts = sub.groupby(id_col).size()
    if counts.empty:
        raise ValueError("No rows match the filter; check filter_col/filter_value.")
    n = counts.mode().iloc[0]  # most common length across ids

    # 3) keep ids with at least n rows and take first n
    good_ids = counts[counts >= n].index
    sub = sub[sub[id_col].isin(good_ids)].copy()
    sub["pos"] = sub.groupby(id_col).cumcount()  # 0..(len-1) within each id
    sub = sub[sub["pos"] < n]
    # 4) pivot to wide
    mat = sub.pivot(index=id_col, columns="pos", values="SUMMARY_VALUE")
    mat = mat.reindex(columns=range(n))  # ensure columns 0..n-1
    mat.columns = [f"{value_col}_{i+1}" for i in range(n)]
    return mat, int(n)

def cluster_fixed_length_matrix(
    mat: pd.DataFrame,
    n_clusters: int = 2,
    random_state: int = 0
):
    """
    If values are numeric -> scale then KMeans.
    If values are non-numeric (categorical) -> factorize per column before scaling.
    """
    X = mat.copy()

    # if any non-numeric column, factorize each column separately
    for c in X.columns:
        if not np.issubdtype(X[c].dtype, np.number):
            X[c] = pd.factorize(X[c])[0]

    scaler = StandardScaler()
    Xs = scaler.fit_transform(X)

    kmeans = KMeans(n_clusters=n_clusters, n_init="auto", random_state=random_state)
    labels = kmeans.fit_predict(Xs)

    # attach cluster labels back to ids
    out = mat.copy()
    out["cluster"] = labels
    return out, kmeans
```
